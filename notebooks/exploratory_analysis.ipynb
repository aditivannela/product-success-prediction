```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of average ratings
sns.histplot(df['avg_review_rating'], bins=20, kde=True)
plt.title("Average Review Rating Distribution")
plt.show()

# Scatter plot: review_count vs avg_review_rating
sns.scatterplot(data=df, x='review_count', y='avg_review_rating', hue='is_successful')
plt.title("Review Count vs Average Rating")
plt.show()
```
